```python
import os
from src import mylib

keyPath=os.path.join(os.getcwd(), 'src', 'key.properties')
key=mylib.getKey(keyPath)
_key=str(key['03_07'])
```

### 2. url 호출
인자를 별도로 넣지 않은 방식

```python
import requests
import re
busstopurl='http://openAPI.seoul.go.kr:8088/'+_key+'/xml/CardBusTimeNew/1/5/202107/7016'
data=requests.get(busstopurl).text

print(data)
```

### 3. XML 파싱 (정규식)
`<태그>(.+?)</태그>` 
* `(.+?)` = 정규식
* 시작 태그와 끝 태그 안에 있는 문자열(정규식)을 가져옴
* `(.+?)` = 문자열
* `(\d+)` = 숫자로 된 문자열
* `\d` = 숫자
* `+` = 한 개 이상이 있다

#### 1) 버스 정류장만 (문자)
BUS_STA_NM

```python
p=re.compile('<BUS_STA_NM>(.+?)</BUS_STA_NM>') # 정규식은 complie을 해야 빠르다
res=p.findall(data)
for item in res:
    print (item)
```

#### 2) 승하차 인원만 (숫자)
*_NUM = 끝부분이 _NUM으로 끝나는 태그 모두

```python
p=re.compile('<.*_NUM>(\d+)</.*_NUM>')
res=p.findall(data)
#print(" ".join(res))
for item in res:
    print (item, end=", ") # connecting numbers with a comma
```

## 코드 통합 (XML 정규식)

```python
%%writefile src/ds_open_cardBusTime.py
# coding: utf-8 # 한글
import os
import mylib # NOTE: do not use 'from src import mylib'
import requests
import re

def doIt():
    keyPath=os.path.join(os.getcwd(), 'src', 'key.properties')
    key=mylib.getKey(keyPath)
    _key=str(key['03_07'])

    busstopurl='http://openAPI.seoul.go.kr:8088/'+_key+'/xml/CardBusTimeNew/1/5/202107/7016'
    data=requests.get(busstopurl).text
    p=re.compile('<BUS_STA_NM>(.+?)</BUS_STA_NM>')
    res=p.findall(data)
    for item in res:
        print (item)

    p=re.compile('<.*_NUM>(\d+)</.*_NUM>')
    res=p.findall(data)
    print(" ".join(res))
if __name__ == "__main__":
    doIt()
```
