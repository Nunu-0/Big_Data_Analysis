### 1. 인증키

```python
import os
from src import mylib

keyPath=os.path.join(os.getcwd(), 'src', 'key.properties')
key=mylib.getKey(keyPath)
```

### 2. 요청인자 (xml)

```python
_url='http://openAPI.seoul.go.kr:8088'
_key=str(key['02_05'])
_type='xml'
_service='CardSubwayTime'
_start_index=1
_end_index=5
_use_mon='202106'
```

### 3. url 호출

```python
import requests

_maxIter=1 # 증가시켜도 상관ㄴㄴ, 반복회수
_iter=0 # 변수 조정
while _iter<_maxIter:
    _api="/".join([_url,_key,_type,_service,str(_start_index),str(_end_index),_use_mon])
    #print _api
    response = requests.get(_api).text
    print(response)
    _start_index+=5
    _end_index+=5
    _iter+=1
```

### 4. xml -> json변환

```python
_type='json' #타입만 json으로 변경한 것
```

```python
_api="/".join([_url,_key,_type,_service,str(_start_index),str(_end_index),_use_mon])
```

```python
data=requests.get(_api).text
print(data)
```

### 5. json 파싱

```python
import json
jsubwayTime = json.loads(data)
```

## 코드 통합(json)

```python
import os
from src import mylib
import requests

keyPath=os.path.join(os.getcwd(), 'src', 'key.properties')
key=mylib.getKey(keyPath)

_url='http://openAPI.seoul.go.kr:8088'
_key=str(key['02_05'])
_type='json'
_service='CardSubwayTime'
_start_index=1
_end_index=5
_use_mon='202106'

_maxIter=2
_iter=0
while _iter<_maxIter:
    _api="/".join([_url,_key,_type,_service,str(_start_index),str(_end_index),_use_mon])
    #print _api
    subwayTime = requests.get(_api).text
    print(subwayTime)
    _start_index+=5
    _end_index+=5
    _iter+=1
```
