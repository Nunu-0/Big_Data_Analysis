## Sympy로 풀기

### 설정

#### 데이터

```python
import numpy as np

x=np.array([1,2,3,4])
y=np.array([6,5,7,10])
#x=np.array([0.5,2.3,2.9])
#y=np.array([1.4,1.9,3.2])
```

```python
x=np.array([x, np.ones(len(x))])  # w0 * x0 + w1 * x1
```

```python
x=x.T
```

```python
x, x.shape
```

####  학습비율

```python
alpha = 0.01
```

#### Weights 초기화

```python
theta=np.ones([x.shape[1]])
```

```python
theta
```

### gradient 

#### h(x)

```python
def h(x, theta):
    return np.dot(x, theta)

h(x,theta)
```

#### 단순 오류
h - y

```python
error=h(x,theta)-y
```

```python
f"simple error h-y: {error}"
```

#### gradient

```python
np.sum(h(x,theta)-y)*(2)/len(x)
```

#### x에 대한 w

```python
np.sum(x.T*(h(x,theta)-y), axis=1)*(2)/len(x)
```

#### dot연산으로 한번에 계산

```python
gradient=np.dot(x.T,error)*2/len(x) # w0, w1
```

```python
f"gradient: {gradient}"
```

### update

```python
theta -= alpha*gradient
```

```python
f"theta: {theta}"
```

### 반복 및 종료

```python
def gradientDescent(x, y, theta, alpha, numIterations):
    """
    new on 20201126
    in:
        x array
        y array
        #alpha=0.01
        #theta=np.array(np.ones([x.shape[1]])) # 2 of (4,2)
        #numIterations=10000
    out: theta weights
    """
    #alpha=0.01
    #theta=np.array(np.ones([x.shape[1]])) # 2 of (4,2)
    #numIterations=10000
    #i=0
    #oldCost=-np.inf
    #cost=np.inf
    for i in range(numIterations):
        #i+=1
        h=np.dot(x, theta)
        error=h-y
        #oldCost=cost
        cost=np.sum((h-y)**2) / len(x) # full MSE
        #cost=np.sum((h-y)**2) / (2 * len(x)) # half MSE
        #gradient=[np.sum(h-y)*2, np.sum((h-y)*2)*2]
        gradient=np.dot(x.T,error)*2/len(x) # w0, w1
        theta-=alpha*gradient
        if (i%100==0):
            print ("iter:{0:5d}\ttheta {1} Cost {2:.5f}".format(i, theta, cost))
    return theta
```

### Python List 데이터의 Gradient Descent

```python
import random

def computeMSE(a,b,x,y):
    """
    new on 20201126
    """
    totalError = 0
    for i in range(0, len(x)):
        totalError += (y[i] - (a + b* x[i])) ** 2
    return totalError / float(len(x))


def gradientDescentL(x,y,alpha,iter):
    """
    new on 20201126
    in:
        x: attribute, 1d float array or list
        y: class, 1d int array
        alpha: learning rate
        iter: n of iterations
    out:
        a, b (weights)
    """
    a=random.random()
    b=random.random()
    #alpha=0.01
    n=len(x)
    for j in range(iter):
        aGradient = 0
        bGradient = 0
        for i in range(n):
            aGradient += (2./n) * (((a + b * x[i])) - y[i])*(1)
            bGradient += (2./n) * (((a + b * x[i])) - y[i])*(x[i])
            #aGradient += (2./n) * (y[i] - ((a + b * x[i])))*(-1)
            #bGradient += (2./n) * (y[i] - ((a + b * x[i])))*(-x[i])
        a = a - (alpha * aGradient)
        b = b - (alpha * bGradient)
        if (j%100==0):
            print ("iter:{0} a={1:.3f} b={2:.3f} AvgError={3:.3f}".format(j,a,b,computeMSE(a,b,x,y)))
    return a, b
```

#### 데이터

```python
#x=np.array([1,2,3,4])
#y=np.array([6,5,7,10])
x=[1,2,3,4]
y=[6,5,7,10]
```

#### 실행

```python
alpha = 0.01
a,b=gradientDescentL(x,y,alpha,5000)
```

#### w출력

```python
print("---> a={0}, b={1} after iterations".format(a,b))
```

#### 예측

### matrix inverse method

#### 데이터

```python
import os

file=open(os.path.join('data','ds_spark_heightweight.txt'))
```

```python
import numpy as np
my=list()
for line in file:
    line=line.split()
    my.append([float(line[1]),float(line[2])])
my=np.array(my)
```

```python
x=my[:,0]
y=my[:,1]
```

```python
x=np.array([x,np.ones(len(x))])
x=x.T
```

```python
x, x.shape
```

```python
y
```

#### w값

```python
w1, w0=np.dot(np.linalg.inv(np.dot(x.T, x)), np.dot(x.T, y))
```

```python
print("w1:{} w0:{}".format(w1,w0))
```

#### yhat

```python
yhat = w0*x[:,1] + w1*x[:,0]
```

#### MSE

```python
np.sum(np.power(y-yhat, 2))/len(x)
```

#### 그래프

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(x[:,0],y)
plt.plot(x[:,0],yhat)
plt.show()
```
