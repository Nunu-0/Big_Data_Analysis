### Sparse Matrix 희소 행렬
0값이 아닌 NZ Nonzero만 저장
```
[1 0 2]
[0 0 3]
[4 5 6]
```
- 행 | 0 | 0 | 1 | 2 | 2 | 2
- 열 | 0 | 2 | 2 | 0 | 1 | 2
- 값 | 1 | 2 | 3 | 4 | 5 | 6

````python
import numpy as np

row = np.array([0, 0, 1, 2, 2, 2])
col = np.array([0, 2, 2, 0, 1, 2])
data = np.array([1, 2, 3, 4, 5, 6])
````

#### scipy

````python
import scipy.sparse as sps

mtx = sps.csc_matrix((data, (row, col)), shape=(3, 3))
````

````python
print (mtx.todense())
````

### Row Matrix
List -> RDD -> Row Matrix

<strong>list<strong/>

````python
p = [[1.0,2.0,3.0],[1.1,2.1,3.1],[1.2,2.2,3.3]]
````

<strong>RDD<strong/>

````python
my = spark.sparkContext.parallelize(p)
````

````python
my.collect()
````

<strong>Row Matrix<strong/>

````python
from pyspark.mllib.linalg.distributed import RowMatrix

rm=RowMatrix(my)
````

````python
print (type(rm))
````

````python
rm.rows.collect()
````

````python
rm.numRows()
````

````python
rm.numCols()
````

### Indexed Row Matrix
순서대로 저장됨

````python
from pyspark.mllib.linalg.distributed import IndexedRow

irRdd = spark.sparkContext.parallelize([
    IndexedRow(1, [3, 1, 2]),
    IndexedRow(2, [1, 3, 2]),
    IndexedRow(3, [5, 4, 3]),
    IndexedRow(4, [6, 7, 4]),
    IndexedRow(5, [8, 9, 2]),
])
````

````python
from pyspark.mllib.linalg.distributed import IndexedRowMatrix

irm = IndexedRowMatrix(irRdd)
````

````python
irm.rows.collect()
````

````python
print(irm.numRows())
print(irm.numCols())
````

### Spark로 RDD 생성
어떻게 구분되어있는지 확인하기

````python
_rdd=spark.sparkContext.textFile(_fsvm)\
    .map(lambda line: [float(x) for x in line.split()])
````

### LabeledPoint 생성
- 0번째 : Label
- 나머지 : feature

````python
from pyspark.mllib.regression import LabeledPoint

_trainRdd0=_rdd.map(lambda line:LabeledPoint(line[0], line[1:]))
````

````python
_trainRdd0.take(1)
````

<strong>구분 + LabelPoint 생성<strong/>

````python
_trainRdd=spark.sparkContext.textFile(_fsvm)\
    .map(lambda line: [float(x) for x in line.split()])\
    .map(lambda p:LabeledPoint(p[0], p[1:]))
````

````python
_trainRdd.take(1)
````

### 정리
데이터 변환 함수 + LabelPoint

````python
def createLP(line):
    p = [float(x) for x in line.split()]
    return LabeledPoint(p[0], p[1:])

_rdd=spark.sparkContext.textFile(_fsvm)
trainRdd = _rdd.map(createLP)

trainRdd.take(1)
````

## TF(Term Frequency) 단어 빈도로 텍스트로 변환
텍스트 변환, 단어 빈도 계산

Hashing으로 자동으로 단어 빈도 계산

````python
wikiRdd3 = spark.sparkContext\
    .textFile(os.path.join("data","ds_spark_wiki.txt"))\
    .map(lambda line: line.split())
````

(로우에서 벡터의 개수, {자동 id : 빈도})

````python
from pyspark.mllib.feature import HashingTF

hashingTF = HashingTF()
tf = hashingTF.transform(wikiRdd3)
tf.collect()
````

## TF-IDF
IDF : 몇 개의 문서에 씌였는지 반대로 계산

훈련데이터를 만들때 사용

````python
from pyspark.mllib.feature import HashingTF, IDF

idf = IDF().fit(tf)
tfidf = idf.transform(tf)
````

````python
tfidf.collect()
````
