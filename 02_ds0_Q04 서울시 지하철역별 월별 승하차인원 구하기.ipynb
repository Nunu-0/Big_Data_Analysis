### 1. 인증키 읽기

```python
import os
import src.mylib

keyPath=os.path.join(os.getcwd(), 'src', 'key.properties')
key=src.mylib.getKey(keyPath)
```

### 2. 요청인자 설정 (json)

```python
base_url = "http://openAPI.seoul.go.kr:8088"

_key = str(key["02_04"])
_type = "json"
_service = "CardSubwayStatsNew"
_start_index = 1
_end_index = 1000
_use_dt = "20151101"
#_maximum_call = 1000
```


### 3. url 만들기 / 호출

```python
params = "/".join([_key, _type, _service, str(_start_index), str(_end_index), _use_dt])
_url = base_url + "/" + params

#_api=os.path.join(_url,_key,_type,_service,str(_start_index),str(_end_index),_use_dt) 
# 이거 사용하면 \로 나옴

print(_url)

# 호출
import requests
response = requests.get(_url)
```

### 4. json 파싱

```python
sub=response.json()
```

```python
print(sub)
```
